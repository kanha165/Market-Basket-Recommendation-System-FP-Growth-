#  Market Basket Analysis using FP-Growth Algorithm

###  Objective:
To find frequent itemsets and association rules using FP-Growth.

###  Dataset Format:
- Each row = transaction
- Columns = items
- NaN = missing item

###  Why FP-Growth?
- Faster than Apriori
- No candidate generation
- Uses FP-Tree

##  Step 1: Import Libraries

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth, association_rules
```

##  Step 2: Load Dataset

```python
data = pd.read_csv("market_basket.csv", header=None)

data.head()
```

##  Step 3: Convert Rows into Transactions

```python
transactions = []

for i in range(len(data)):
    transaction = []
    
    for item in data.iloc[i]:
        if pd.notna(item):
            transaction.append(str(item).strip().lower())
    
    transactions.append(transaction)

transactions[:5]
```

##  Step 4: One-Hot Encoding

```python
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)

df = pd.DataFrame(te_array, columns=te.columns_)

df.head()
```

##  Step 5: Apply FP-Growth

```python
frequent_itemsets = fpgrowth(df, min_support=0.02, use_colnames=True)

frequent_itemsets.sort_values(by="support", ascending=False).head(10)
```

##  Step 6: Generate Rules

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.3)

rules.sort_values(by="lift", ascending=False).head(10)
```

##  Step 7: Strong Rules

```python
strong_rules = rules[(rules['confidence'] > 0.4) & (rules['lift'] > 1)]

strong_rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
```

```python
# Save rules
rules.to_csv("fp_rules.csv", index=False)

# Save itemsets (optional)
frequent_itemsets.to_csv("fp_itemsets.csv", index=False)

print(" Rules & Itemsets saved successfully!")
```

```python
import pandas as pd
import ast

rules = pd.read_csv("fp_rules.csv")

# Clean frozenset string
def clean_frozenset(x):
    x = x.replace("frozenset(", "").replace(")", "")
    return list(ast.literal_eval(x))

rules['antecedents'] = rules['antecedents'].apply(clean_frozenset)
rules['consequents'] = rules['consequents'].apply(clean_frozenset)

print(" Fixed and Loaded!")
```
